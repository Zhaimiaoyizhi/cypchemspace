# cypchemspace Demo

This notebook runs the teaching workflow from a small CYP substrate table to fingerprints, embedding, enrichment statistics, and a core figure.

```python
from pathlib import Path
from cypchemspace._compat import avoid_windows_wmi_platform_probe
avoid_windows_wmi_platform_probe()
import pandas as pd

from cypchemspace import load_substrate_table, make_morgan_fingerprint_matrix, knn_label_enrichment
from cypchemspace.embedding import compute_embedding
from cypchemspace.visualize import plot_embedding

data_path = Path('example_data/cyp_substrates_demo.csv')
out_dir = Path('../results/notebook_demo')
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
table = load_substrate_table(data_path)
table.groupby('label').size()
```

```python
fingerprints, valid_table = make_morgan_fingerprint_matrix(table)
embedding = compute_embedding(fingerprints, method='umap')
embedded = pd.concat([valid_table.reset_index(drop=True), embedding], axis=1)
embedded.head()
```

```python
stats = knn_label_enrichment(fingerprints, embedded['label'], positive_label='mydb', k=3, n_permutations=99, random_state=42)
stats
```

```python
embedded.to_csv(out_dir / 'embedding.csv', index=False)
pd.DataFrame([stats]).to_csv(out_dir / 'summary.csv', index=False)
plot_embedding(embedded, out_dir / 'cypchemspace_umap.png')
```
